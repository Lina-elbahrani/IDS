```python
import pandas as pd
import numpy as np
import joblib

from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.metrics import confusion_matrix, classification_report, f1_score
```

```python
# ===============================
# 1. Chargement des données
# ===============================
train_df = pd.read_csv("KDDTrain+.txt", header=None)
test_df  = pd.read_csv("KDDTest+.txt", header=None)
```

```python
# ===============================
# 2. Noms des colonnes (41 features + label + difficulty)
# ===============================
columns = [
    'duration','protocol_type','service','flag','src_bytes','dst_bytes',
    'land','wrong_fragment','urgent','hot','num_failed_logins','logged_in',
    'num_compromised','root_shell','su_attempted','num_root',
    'num_file_creations','num_shells','num_access_files','num_outbound_cmds',
    'is_host_login','is_guest_login','count','srv_count','serror_rate',
    'srv_serror_rate','rerror_rate','srv_rerror_rate','same_srv_rate',
    'diff_srv_rate','srv_diff_host_rate','dst_host_count','dst_host_srv_count',
    'dst_host_same_srv_rate','dst_host_diff_srv_rate',
    'dst_host_same_src_port_rate','dst_host_srv_diff_host_rate',
    'dst_host_serror_rate','dst_host_srv_serror_rate',
    'dst_host_rerror_rate','dst_host_srv_rerror_rate',
    'label','difficulty'
]

train_df.columns = columns
test_df.columns = columns
```

```python
import pandas as pd

# Option pour voir toutes les colonnes sans coupure
pd.set_option('display.max_columns', None)

print("=== Aperçu complet (Train) ===")
display(train_df.head())  # 'display' rend le tableau plus joli dans Jupyter

print("\n=== Aperçu complet (Test) ===")
display(test_df.head())
```

```python
# ===============================
# 3. Label binaire (normal = 0, attaque = 1)
# ===============================
train_df['label'] = train_df['label'].apply(lambda x: 0 if x == 'normal' else 1)
test_df['label']  = test_df['label'].apply(lambda x: 0 if x == 'normal' else 1)

# ===============================
# 4. Suppression colonne difficulty
# ===============================
train_df.drop(columns=['difficulty'], inplace=True)
test_df.drop(columns=['difficulty'], inplace=True)

# ===============================
# 5. Encodage des colonnes catégorielles
# ===============================
categorical_cols = ['protocol_type', 'service', 'flag']

encoders = {}

for col in categorical_cols:
    le = LabelEncoder()
    train_df[col] = le.fit_transform(train_df[col])
    test_df[col]  = le.transform(test_df[col])
    encoders[col] = le

joblib.dump(encoders, "label_encoders.pkl")


# ===============================
# 6. Séparation X / y
# ===============================
X_train = train_df.drop('label', axis=1)
y_train = train_df['label']

X_test  = test_df.drop('label', axis=1)
y_test  = test_df['label']

# ===============================
# 7. Normalisation
# ===============================
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test  = scaler.transform(X_test)

joblib.dump(scaler, "scaler.pkl")
```

```python
# ===============================
# 8. Random Forest
rf = RandomForestClassifier(
    n_estimators=200,   # On passe de 100 à 200 arbres
    max_depth=None,     # On enlève la limite de profondeur !
    min_samples_split=5, # Pour éviter d'apprendre trop par cœur
    random_state=42,
    n_jobs=-1           # Utilise tous les cœurs de votre PC pour aller vite
)
rf.fit(X_train, y_train)
rf_pred = rf.predict(X_test)

print("\n=== RANDOM FOREST ===")
print(confusion_matrix(y_test, rf_pred))
print(classification_report(y_test, rf_pred))
print("F1-score:", f1_score(y_test, rf_pred))

joblib.dump(rf, "random_forest_ids.pkl")
```

```python
# ===============================
# 9. SVM
# ===============================
svm = SVC(
    C=1,
    kernel='rbf',
    gamma='scale'
)

svm.fit(X_train, y_train)
svm_pred = svm.predict(X_test)

print("\n=== SVM ===")
print(confusion_matrix(y_test, svm_pred))
print(classification_report(y_test, svm_pred))
print("F1-score:", f1_score(y_test, svm_pred))

joblib.dump(svm, "svm_ids.pkl")
```

```python
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Input, Dense
from tensorflow.keras.callbacks import EarlyStopping
import numpy as np
```

```python
# Autoencoder entraîné UNIQUEMENT sur le trafic normal
X_train_normal = X_train[y_train == 0]

input_dim = X_train.shape[1]

input_layer = Input(shape=(input_dim,))
encoded = Dense(64, activation='relu')(input_layer)
encoded = Dense(32, activation='relu')(encoded)

decoded = Dense(64, activation='relu')(encoded)
decoded = Dense(input_dim, activation='linear')(decoded)

autoencoder = Model(inputs=input_layer, outputs=decoded)
autoencoder.compile(
    optimizer='adam',
    loss='mse'
)

autoencoder.summary()
```

```python
early_stop = EarlyStopping(
    monitor='val_loss',
    patience=3,
    restore_best_weights=True
)

autoencoder.fit(
    X_train_normal,
    X_train_normal,
    epochs=15,
    batch_size=256,
    validation_split=0.1,
    callbacks=[early_stop],
    verbose=1
)
```

```python
# Erreur de reconstruction sur le train normal
reconstructions = autoencoder.predict(X_train_normal)
train_loss = np.mean(np.square(reconstructions - X_train_normal), axis=1)

# Seuil (95%)
threshold = np.percentile(train_loss, 95)
print("Threshold:", threshold)

# Reconstruction sur test
test_reconstructions = autoencoder.predict(X_test)
test_loss = np.mean(np.square(test_reconstructions - X_test), axis=1)

# Prédiction : 1 = attaque, 0 = normal
ae_pred = (test_loss > threshold).astype(int)
import json

with open("threshold.json", "w") as f:
    json.dump({"threshold": float(threshold)}, f)

```

```python
from sklearn.metrics import confusion_matrix, classification_report, f1_score

print("\n=== AUTOENCODER ===")
print(confusion_matrix(y_test, ae_pred))
print(classification_report(y_test, ae_pred))
print("F1-score:", f1_score(y_test, ae_pred))

autoencoder.save("autoencoder_ids.h5")
```
